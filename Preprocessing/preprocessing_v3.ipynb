```python
from datasets import DatasetDict
import re
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder

# Load LegalBERT for STF-IDF
tokenizer = AutoTokenizer.from_pretrained("nlpaueb/legal-bert-base-uncased")
model = AutoModel.from_pretrained("nlpaueb/legal-bert-base-uncased")
model.eval()

# ------------------ 1. Abbreviation Expansion ------------------


def expand_abbreviations(text):
    for pattern, repl in legal_abbrev.items():
        text = re.sub(pattern, repl, text)
    return text

# ------------------ 2. Metadata Removal ------------------
def remove_metadata(text):
    text = re.sub(r"^(Decided on|Argued on|Filed:|Docket No).*", "", text, flags=re.MULTILINE)
    return re.sub(r"\n+", "\n", text.strip())

# ------------------ 3. Paragraph Segmentation ------------------
def segment_paragraphs(text):
    return [p.strip() for p in text.split('\n') if len(p.strip()) > 30]

# ------------------ 4. Paragraph Filtering ------------------
def embed(texts):
    inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt", max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state[:, 0, :].numpy()  # CLS token

def stf_idf_filter(paragraphs, top_k=3):
    doc_embedding = embed([" ".join(paragraphs)])[0]
    para_embeddings = embed(paragraphs)
    similarities = np.dot(para_embeddings, doc_embedding) / (
        np.linalg.norm(para_embeddings, axis=1) * np.linalg.norm(doc_embedding) + 1e-10)
    top_idx = np.argsort(similarities)[-top_k:]
    return [paragraphs[i] for i in sorted(top_idx)]

# ------------------ 5. Wrap as HuggingFace .map() preprocessing ------------------
def preprocess(example):
    text = expand_abbreviations(example["text"])
    text = remove_metadata(text)
    paragraphs = segment_paragraphs(text)
    
    if len(paragraphs) == 0:
        return {"text": ""}
    
    selected = stf_idf_filter(paragraphs, top_k=3)
    return {"text": " ".join(selected)}

# ------------------ 6. Apply using .map() ------------------
processed_dataset = dataset.map(preprocess, batched=False)
```
